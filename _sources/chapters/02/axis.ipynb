```python
import numpy as np
import matplotlib.pyplot as plt 
```

scatter 함수와 함께 사용할 수 있는 다양한 시각화 함수들을 살펴봅니다. 
먼저 scatter 함수의 옵션인 color와 marker를 확인해봅시다. 어떤 color와 marker가 사용가능한지 입력할 수 있는 값을 찾아보세요.

```python
t = np.arange(0., 5., 0.2)
fig, ax = plt.subplots(figsize=(10,5))
ax.scatter(t,t, color ='red', marker = '.')
ax.set_title('scatter plot')
```

그림에 grid를 추가해봅니다. grid 함수를 통해 쉽게 구현할 수 있습니다.

```python
# grid의 추가
t = np.arange(0., 5., 0.2)
fig, ax = plt.subplots(figsize=(10,5))
ax.scatter(t,t, color ='red', marker = '.')
ax.grid(True)
```

Grid 를 표시하였습니다. grid의 간격을 조정할 수 있을까요? grid의 간격와 관련한 정보는 ticks에 있습니다. x축에 대한 것은 ***xticks***  y 축에 대한 것은 ***yticks***에 있습니다. 

grid 간격을 조정하는 부분을 찾아보고 해석해보세요.

```python
#grid 조정
t = np.arange(0., 5., 0.2)
fig, ax = plt.subplots(figsize=(10,5))
ax.scatter(t,t, color ='red', marker = '.')
ax.grid(True)
ax.set_xticks(np.arange(0,5,0.5))
ax.set_title('scatter plot')
```

*legend*도 쉽게 추가할 수 있습니다. 

```python
t = np.arange(0., 5., 0.2)
fig, ax = plt.subplots(figsize=(10,5))
ax.scatter(t,t, color ='red', marker = '.', label='red_p')
ax.grid(True)
ax.set_xticks(np.arange(0,5,0.5))
ax.set_title('scatter plot')
ax.legend()
```

*legend*의 위치를 변경하기 위해서는 *legend* 함수내에 loc 옵션을 변경하면 됩니다. 위치값은 아래와 같습니다. 
- upper left', 'upper right', 'lower left', 'lower right'
- best, center, center right

```python
t = np.arange(0., 5., 0.2)
fig, ax = plt.subplots(figsize=(10,5))
ax.scatter(t,t, color ='red', marker = '.', label='red_p')
ax.grid(True)
ax.set_xticks(np.arange(0,5,0.5))
ax.set_title('scatter plot')
ax.legend(loc='lower right')
```

하나의 그림(axes)에 그림으 그리는 함수를 반복해서 적용하면 결과물이 겹쳐져서 표현됨을 앞서 살펴보았습니다. 다시 한번 확인해보겠습니다. 

```python
t = np.arange(0., 5., 0.2)
fig, ax = plt.subplots(figsize=(10,5))
ax.scatter(t,t, color ='red', marker = '.', label="linear")
ax.scatter(t,t**2, color ='lightblue', marker = '+', label='quadratic')
ax.scatter(t,t**3, color ='green', marker = '^', label = 'cubic')
ax.legend(loc='upper left')
```

히스토그램을 그려보겠습니다. 히스토그램은 연속형 자료의 확률밀도함수에 대한 시각화 결과물로 bins 옵션을 이용하여 계급구간의 개수(bin의 수)를 정하는 것이 매우 중요합니다. 일반적으로 자료의 수가 $n$일때 계급구간의 수를 $\sqrt{n}$이하인 최대자연수로 잡습니다. 



아래 예제에서 다음을 확인하세요. (여기서는 subplots을 사용하지 않고 plt.hist()로 바로 그렸습니다.)
- bins: bin의 개수
- density: 히스토그램의 높이를 상대도수로 계산 여부 (default는 false로 되어 있으나, 히스토그램의 높이를 확률밀도로 해석하기 위해서는 true로 설정하여야 함)
- facecolor: 면의 색깔

```python
x = np.random.normal(0,1,size =2500)
n, bins, patches = plt.hist(x, bins=50, density=True, facecolor='tab:green', alpha=0.75)
```

추가로 아래 그림에 여러 속성값을 정하는 절차를 확인하세요
- xlabel, ylabel: 축 이름 변경
- title: 제목변경
- text: 문자의 표시
- xlim, ylim: 축의 범위 지정

```python
n, bins, patches = plt.hist(x, bins=50, density=True, facecolor='tab:green', edgecolor = 'white', alpha=0.75)
plt.xlabel('x')
plt.ylabel('Probability')
plt.title('standard normal')
plt.text(-0.5, .1, r'$\mu=0,\ \sigma=1$')
plt.xlim(-4, 4)
plt.ylim(0, 0.5)
plt.grid(True)
```

위 시각화 결과물은 실제로 표준정규분포를 따르는 자료의 개형을 잘 나타내고 있습니다. 하지만, 
같은 자료에 대해서 bin의 수를 너무 작게 잡거나 너무 크게 잡을때 히스토그램의 개형이 원래 자료의 분포를 잘 표현하지 못할 수 있습니다.아래 예를 확인해보세요.

```python
plt.hist(x, bins=2, density=True, facecolor='tab:green', alpha=0.75)
plt.title('standard normal')
```

```python
plt.hist(x, bins=2500, density=True, facecolor='tab:green', alpha=0.75)
plt.title('standard normal')
```

여기서는 subplots을 이용하여 히스토그램을 그리는 방법을 확인해보겠습니다. 같은 axes에 두개의 히스토그램을 그리는 예제입니다. 

```python
# 동시비교: 2개 데이터의 표시 
a = np.random.normal(0, 3, 1000)
b = np.random.normal(2, 4, 900)
fig, ax = plt.subplots()
ax.hist(a, 50, density = True, alpha = 0.5, label=r'N(0,$3^2$)')
ax.hist(b, 50, density = True, alpha = 0.5, label=r'N(2,$4^2$)')
ax.legend(loc='upper left')
ax.set_title("two histogramss")
```

히스토그램을 수평 수직으로 따로 그릴 수 있을 것입니다. 주의할 점은 시각적으로 용이한 비교를 위해 x축의 길이를 같게 맞추어 주는 것입니다. (x축길이를 같게 설정하지 않으면 어떤 점이 불편한지 알아봅시다)

```python
# 수평비교
fig, ax = plt.subplots(1,2)
ax[0].hist(a, 50, density = True, alpha = 0.5, label=r'N(0,$3^2$)')
ax[1].hist(b, 50, density = True, alpha = 0.5, label=r'N(2,$4^2$)')
ax[0].legend(loc='upper left')
ax[1].legend(loc='upper left')
```

```python
# 수직비교1
fig, ax = plt.subplots(2,1)
ax[0].hist(a, 50, density = True, alpha = 0.5, label=r'N(0,$3^2$)')
ax[1].hist(b, 50, density = True, alpha = 0.5, label=r'N(2,$4^2$)')
ax[0].legend(loc='upper left')
ax[1].legend(loc='upper left')
```

```python
# 수직비교2
fig, ax = plt.subplots(2,1, figsize = (5,10))
ax[0].hist(a, 50, density = True, alpha = 0.5, label=r'N(0,$3^2$)')
ax[1].hist(b, 50, density = True, alpha = 0.5, label=r'N(2,$4^2$)')
ax[0].legend(loc='upper left')
ax[1].legend(loc='upper left')
ax[0].set_xlim(-10,15)
ax[1].set_xlim(-10,15)
```

여기서는 boxplot을 그려보겠습니다. boxplot은 연속형 변수의 기초통계량을 요약해주는 유용한 시각화 도구입니다. boxplot을 통해서 중앙값, 1분위수, 3분위수, IQR(inter quartile range), 이상치 등을 파악할 수 있습니다.

boxplot은 하나의 변수에 대해서 단독으로 사용하기 보다는 여러 개의 연속형 변수 (다른 모집단의 변수)를 동시에 비교할 때 많이 사용합니다. 

boxplot 함수에 리스트의 형태로 자료가 들어가고 labels를 문자열로 입력함을 살펴보세요.

```python
a = np.random.normal(0, 1, 1000)
b = np.random.normal(5, 1, 200)
c = np.random.normal(2, 2, 200)
d = np.random.normal(1, 3, 200)
fig, ax = plt.subplots()
ax.boxplot([a,b,c,d], labels=["A","B", "C", "D"], )
ax.set_title("boxplot")
```

boxplot은 연속형 변수의 자료 요약값을 보여줄 수 있는 매우 유용한 도구지만 복잡한 분포의 형태를 나타낼 수는 없습니다. 특별히 혼합형 분포를 가지는 자료분석의 예를 살펴보겠습니다. 먼저 이러한 분포를 하나 만들어겠습니다. 

```python
# mixture 분포
a = np.random.normal(0, 1, 1000)
b = np.random.normal(5, 1, 200)
x = np.concatenate([a,b])
fig, ax = plt.subplots()
ax.hist(x, bins = 40)
ax.set_title('mixture dist')
```

두 개의 분포가 혼합된 형태의 자료로 보입니다. 특히 솟아오른 두개의 봉우리 즉, 자료가 많이 모여 있는 곳이 두 군데가 보입니다. 이 자료를 boxplot을 통해 특성을 파악하면 어떻게 될까요?

```python
fig, ax = plt.subplots()
ax.boxplot(x, labels='x')
ax.set_title('mixture dist')
```

boxplot은 혼합형 분포의 특징을 잘 나타내지 못하고 있습니다. 이러한 boxplot의 단점을 보완하기 위해 만들어진 시각화 도구가 violin plot입니다. 

violin plot은 boxplot의 정보를 모두 담으면서 밀도함수의 추정값(부드러운 히스토그램)을 같이 표시함으로써 boxplot이 나타내지 못한는 더 많은 정보를 담고 있습니다. 물론 정보 양의 크기와 간결성 사이에서 우리는 신중한 선택을 해야할 것입니다. 

```python
fig, ax = plt.subplots()
ax.violinplot(x)
ax.set_title('mixture dist')
```

```python
fig, ax = plt.subplots()
ax.violinplot([x,a,b,c,d])
ax.set_title("violin plot")
```

violon plot 마다 label 값을 할당하기 위해서 set_xtickslabels 을 이용한다.

```python
# labeling의 문제점 해결
fig, ax = plt.subplots()
ax.violinplot([x,a,b,c,d])
ax.set_title('mixture dist')
ax.set_xticks([1,2,3,4,5])
ax.set_xticklabels(["mixture", 'a', 'b', 'c','d'])
```
