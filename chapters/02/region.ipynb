```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

***스타일의 선택***

- 스타일이란 시각화에 사용되는 열 가지 시각적 객체의 속성(예: 색깔, 크기, 모양 등)의 정해진 값들을 의미합니다. 좋은 스타일은 이러한 시각 개체 속성을 미리 잘 만들어 정보가 잘 전달되고 아름다운 시각화가 가능하게 만들어둔 시각속성 템플릿입니다. 아래와 같이 style.use()를 이용하여 스타일을 지정할 수 있습니다.

```python
#plt.style.use("ggplot") 
# 속성을 변경해서 다시 그려보세요.
plt.style.use("default")
# 회귀모형의 시각화
np.random.seed(1)
x  = np.arange(0,1, 0.001)
y  = x + np.random.normal(0,1, size = len(x))
x
plt.plot(x)
```

아래는 R언어로 시각화에 널리 사용하는 'ggplot' 패키지에서 제공하는 스타일로 위 그래프를 다시 그린 결과다. 아래와 같이 스타일이 적용되면 현재 실행되는 

```python
plt.style.use("ggplot")
plt.plot(x)
```

```python
plt.plot(x,y,'.')
```

아래 그림은 alpha 값을 0.2 로 변경하여 만든 결과물입니다. 
- alpha는 불투명도의 크기를 나타내는 것으로 0과 1의 값을 가지며 기본적으로 1의 값을 가집니다.
- 시각객체를 투명하게 만들기 위해서는 0에 가까운 값을 주면 됩니다.

```python
plt.plot(x,y,'.', alpha = 0.2)
```

다음으로 축의 영역을 정하는 방법입니다. 그림을 그리고 이어서 ylim, xlim을 실행하면 입력값에 맞는 축 영역의 크기로 변경됩니다. 기본적으로는 데이터의 최대, 최소값에 적당한 값을 더하고 빼준 값이 축 영역의 크기가 됩니다.

```python
plt.plot(x,y,'.', alpha = 0.2)
plt.ylim(-10,10)
```

지금까지는 pyplot에서 시각화 함수 (plt.plot, plt.hist 등)의 함수로 직접 그래프를 그리는 연습을 하였습니다.

여기서는 시각화 방법을 조금 더 프로그램 구조를 직접 활용할 수 있도록 코드를 작성하는 방법을 확인해보겠습니다.

***plt.subplots*** 는 시각화를 위한 두 가지 객체 (figure, axes)를 생성합니다.
- figure: 그림의 레이아웃, 시각화 속성이 표현되는 최상위 객체
- axes: 그림을 표현하는 객체인 Axis, Tick, Line2D, Text, Polygon, 좌표계

사용법은 다음과 같습니다.
- subplots 로 figure 와 axes를 생성합니다.
- axes가 가지고 있는 method인 scatter(산점도), hist(히스토그램), boxplot(상자그림)등을 이용하여 시각화합니다.

```python
x = np.random.uniform(size=10); y = np.random.uniform(size=10)
fig, ax = plt.subplots()
ax.scatter(x,y)
```

다음으로 subplots 에 nrows, ncols 값을 입력하여 2행 1열 그림의 레이아웃을 그려봅니다. 그림의 크기는 figsize 옵션을 이용하여 지정하였습니다. 여기서 레이아웃 별로 axes 객체를 따로 다루기 위해서 tuple 형태로 ax1, ax2 변수를 지정하였습니다.

```python
fig, (ax1, ax2) = plt.subplots(2,1, figsize = (2,4))
ax1.scatter(x,y, color ='blue')
ax2.scatter(y,x, color ='red')
```

axes 에는 그림의 결과물을 변형할 수 있는 많은 함수들이 있습니다. 이 함수 이름에 규칙이 있는데 다음과 같습니다. 
- set or get
- xlabel, ylabel, title, xtick..
- major, minor

아래 예에서는 scatter 함수에서 다양한 그림속성을 변경하는 예를 보여주고 있습니다. 이 예가 보여주는 것은 다음과 같습니다.
- 시각화 속성으로 색상, 크기를 이용하여 2차원 이상의 데이터를 산점도로 표시할 수 있다.
- 제목달기(set_title)
- 축이름 바꾸기(set_xlabel, set_ylabel)

아래 예에서 위 사항을 확인해봅시다.

```python
np.random.seed(1)  # seed the random number generator.
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

fig, ax = plt.subplots(figsize= (10,5))
# c: color, s: size
ax.scatter(x = data['a'], y = data['b'], c = data['c'], s = data['d'])
ax.set_title("main plot")
ax.set_xlabel("x")
ax.set_ylabel("y")
```

여러개의 그림을 그릴 때 axes를 index로 관리할 수도 있습니다. 아래 예에서 ax[]구문을 살펴보세요. 그리고 facecolor, edgecolor 옵션도 알아두면 유용합니다.

```python
# color의 속성을 변경할 수 있는 대상: facecolor, edgecolors
fig, ax = plt.subplots(1,2, figsize = (10,5))
ax[0].scatter(x,y,s = 50, facecolor = 'red', edgecolors ='blue')
ax[1].scatter(y,x,s = 50, facecolor = 'orange', edgecolors ='red')
```

```python
# color의 속성을 변경할 수 있는 대상: facecolor, edgecolors
fig, ax = plt.subplots(2,2, figsize = (10,10))
ax[0,0].scatter(x,y,s = 50, facecolor = 'red', edgecolors ='blue')
ax[0,1].scatter(y,x,s = 50, facecolor = 'orange', edgecolors ='red')
ax[1,0].scatter(y,x,s = 5, facecolor = 'red', edgecolors ='blue')
ax[1,1].scatter(y,x,s = 5, facecolor = 'orange', edgecolors ='red')
```

다음 예는 각은 그래픽 객체에 plot 함수를 두번 적용하면 일어나는 일을 보여줍니다. 이 결과를 보면 axes 객체에 어떤 함수가 중복하여 적용됨을 알 수 있습니다. 

한편 함수 적용시 label을 달아놓아 결과값을 구분할 수 있으며 이를 범례함수(legend)가 참조함을 확인할 수 있습니다.

```python
# legend를 붙이기 위해서 label을 설정하는 것을 확인
fig, ax = plt.subplots(figsize = (10,5))
ax.plot(x, color='red', label = 'Korea')
ax.plot(y, color = 'blue', label = 'US')
ax.legend()
```
