## 스타일의 선택

- 스타일이란 ...
아래와 같이 style.use()를 이용하여 스타일을 지정할 수 있다. 

```python

#plt.style.use("ggplot")
plt.style.use("default")
#%%
# (1) 회귀모형의 시각화
np.random.seed(1)
x  = np.arange(0,1, 0.001)
y  = x + np.random.normal(0,1, size = len(x))
x
plt.plot(x)
```

아래는 R언어로 시각화에 널리 사용하는 'ggplot' 패키지에서 제공하는 스타일로 위 그래프를 다시 그린 결과다. 아래와 같이 스타일이 적용되면 현재 실행되는 

```python
plt.style.use("ggplot")
plt.plot(x)
```
