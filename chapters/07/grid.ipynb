```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
# seaborn.PairGrid (iris dataset)
## pairplot 활용
iris = sns.load_dataset("iris")
sns.pairplot(data=iris)
sns.pairplot(iris, hue="species")
## PairGrid 활용
sns.PairGrid(data=iris) # pairplot과 다르게 plot이 나오지 않음
### 해결방안 all the grids plot scatter
g = sns.PairGrid(iris)
g.map(plt.scatter)

# seaborn.FacetGrid (tips dataset)
tips = sns.load_dataset("tips")
g = sns.FacetGrid(tips, col="time", row="smoker") #plot이 나오지 않음
## 해결방안
g = sns.FacetGrid(tips, col="time", row="smoker")
g.map(plt.hist, "total_bill")

# sns.JointGrid (tips dataset)
g = sns.JointGrid(tips, x="total_bill", y="tip") # plot이 나오지 않음
## 해결방안
g = sns.JointGrid(tips, x="total_bill", y="tip")
g.plot(sns.scatterplot, sns.histplot)
```

```python
# Multiple relationships with facets
tips = sns.load_dataset("tips")
## Chater06에서 배운 방법
sns.relplot(
    data=tips,
    x="total_bill", y="tip", hue="smoker", style="sex",
)
## Showing multiple relationships with facets
sns.relplot(
    data=tips,
    x="total_bill", y="tip", hue="smoker", col="sex",
)
## facet_grid를 이용한 산점도
g = sns.FacetGrid(tips, col="sex", hue="smoker")
g.map(sns.scatterplot, "total_bill", "tip", alpha=.7)
g.add_legend()

with sns.axes_style("white"):
    g = sns.FacetGrid(tips, row="sex", col="smoker", margin_titles=True, height=2.5)
g.map(sns.scatterplot, "total_bill", "tip", color="#334488")
g.set_axis_labels("Total bill (US Dollars)", "Tip")
g.set(xticks=[10, 30, 50], yticks=[2, 6, 10])
g.figure.subplots_adjust(wspace=.02, hspace=.02)
```

```python
sns.catplot(data=tips, x="sex", y="total_bill", hue="day", kind="bar")
## Showing multiple relationships with facets
sns.catplot(data=tips, x="sex", y="total_bill", col="day", kind="bar")
## facet_grid를 이용한 barplot
g = sns.FacetGrid(tips, col="day", height=4, aspect=.5)
g.map(sns.barplot, "sex", "total_bill", order=["Male", "Female"])
```

```python
##################################################### 7-3-2 ###########################################################
# sinplot 정의
def sinplot(n=10, flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, n + 1):
        plt.plot(x, np.sin(x + i * .5) * (n + 2 - i) * flip)
sinplot()

# Seaborn figure styles
sns.set_style("whitegrid")
sinplot()

sns.set_style("dark")
sinplot()

sns.set_style("white")
sinplot()

sns.set_style("ticks")
sinplot()
```

```python
# Removing axes spines
sinplot()
sns.despine()

sinplot()
sns.despine(left=True)

# Scaling plot elements
sns.set_theme()

sns.set_context("paper")
sinplot()

sns.set_context("notebook")
sinplot()

sns.set_context("talk")
sinplot()

sns.set_context("poster")
sinplot()

sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 2.5})
sinplot()
```
